# Loading the data

```python
data=pd.read_csv('fifa.csv',index_col="Date")

data
```

# Examining the data

```python
data.head()
```

# Plotting data with seaborn


```python
# set the width and height of the figure

plt.figure(figsize=(15,15))

plt.title(" Fifa Dataset ",fontsize=30)
sns.set_style('dark')
#line chart

sns.lineplot(data=data)


```

```python
#list(data.columns)--> list of specific columns

```

## Barchart

```python
plt.figure(figsize=(10,6))
plt.title('Argentina(ARG)')


sns.barplot(x=data.index,y=data['ARG'])
```

## HeatMap

```python
plt.figure(figsize=(20,20))
plt.title("Fifa Dataset")
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
sns.heatmap(data=data,annot=True,annot_kws={"size":18},)
#annot true shows the values on the heatmap

```

## Scatter plot

```python
sns.scatterplot(x=data['FRA'],y=data['ARG'])
```

```python
sns.regplot(x=data['ESP'],y=data['ARG'])
```

## Histogram

```python
sns.histplot(data['ARG'])
```

```python
sns.kdeplot(data['ARG'],shade=True)
sns.set_style('dark')
```

# Distribution plot

```python
sns.displot(data['ARG'])
```

# Box Plot

```python
sns.boxplot(x ='ESP', y = 'ARG', 
                data = data)
```
